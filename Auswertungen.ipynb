# Auswertung

## Deep Q-Learning

### Pong-v0

```python
import matplotlib.pyplot as plt
import pandas as pd

path = "Deep Q-Network/Results/Pong/DQN_Pong-v0_Auswertung.csv"
results = pd.read_csv(path)
```

```python
plt.figure(figsize=(15, 5))
plt.scatter(x=range(0, results['Rewards'].size), y=results['Rewards'], s=3, color="cadetblue", label="Rewards")
plt.plot(results['Mean Reward'], color="red", label="Mean Reward", linewidth=3)
plt.xlabel("Episoden", fontsize=20)
plt.ylabel("Rewards", fontsize=20)
plt.legend()
plt.show()
```
